# Comparision between highest and lowest markets and their commodities

```python
p['price_variation'] = p['max_price'] - p['min_price']
top_variation = p.nlargest(10, 'price_variation')

print("Top 10 Commodities with Highest Price Variation:\n", top_variation[['commodity', 'min_price', 'max_price', 'price_variation']])
```

# price diiference detween different min and max prices and their variation

```python
state_price_fluctuation = p.groupby('state').agg({'min_price': 'min', 'max_price': 'max'})
state_price_fluctuation['price_fluctuation'] = state_price_fluctuation['max_price'] - state_price_fluctuation['min_price']
state_price_fluctuation = state_price_fluctuation.sort_values(by='price_fluctuation', ascending=False)

print("States with Highest Price Fluctuation:\n", state_price_fluctuation.head(10))
print("\nStates with Lowest Price Fluctuation:\n", state_price_fluctuation.tail(10))
```

The state-wise price distribution further emphasizes regional price differences, which may be attributed to factors such as climate conditions, transportation costs, and local market dynamics.
